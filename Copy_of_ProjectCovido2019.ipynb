```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv('covid19.csv')
co = pd.read_csv('/content/CountryAndContinent.csv')

del df["Unnamed: 0"]
df = df.head(221)
#df.fillna(, inplace = True)
# df["TotalCases"].astype(int)
missing_col = ["NewCases","TotalDeaths","TotalCases","Country,Other"]
for col in missing_col:
    df[col] = df[col].fillna(0.0)

# replacing the totalcases column data from str to float
u = len(df["TotalCases"])
for i in range(0,u):
    # x = df.loc[i,'TotalCases']
    # type(x)
    # float(x.replace(',', ''))
    # df.replace(df.loc[i,'TotalCases'],x)
    df.loc[i,'TotalCases'] = float(df.loc[i,'TotalCases'].replace(',', ''))
    
# replacing the newcases column data from str to float
k = len(df["NewCases"])
for i in range(0,k):
    df.loc[i,'NewCases'] = float(str(df.loc[i,'NewCases']).replace(',', ''))

#replacing the " " str to "0" in totaldeath column
for i in range (0,221):
    if df.loc[i,"TotalDeaths"] == " ":
      df.loc[i,"TotalDeaths"] = df.loc[i,"TotalDeaths"].replace(" ", "0")

#replacing the totaldeath data from str to float
df.loc[199,"TotalDeaths"]
p = len(df["TotalDeaths"])
for i in range(0,p):
    df.loc[i,"TotalDeaths"] = float(df.loc[i,"TotalDeaths"].replace(',', ''))

a = df.merge(co , on = "Country,Other")


a.rename(columns = {"NewCases":"New Cases" , "TotalDeaths":"Total Deaths" , "TotalCases" : "Total Cases" , "Country,Other" : "Country"}, inplace = True)
print(a.to_string())


#df
#type(df.loc[0,'NewCases'])
```
